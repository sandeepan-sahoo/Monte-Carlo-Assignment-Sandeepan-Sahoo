```python
import random
import math
import matplotlib.pyplot as plt
#Inversion
def inversion(a):
    u = random.uniform(0,1)
    return math.atan(math.sqrt(a)*math.tan(2*math.pi*math.sqrt(a)*u))
x = []
y = []
for i in range(10000):
    x.append(i)
    y.append(inversion(0.5))
plt.plot(x,y)
plt.xlabel("Number of try")
plt.ylabel("Random number")
plt.title("Inversion plot at a=0.5")
plt.show()
x = []
y = []
for i in range(10000):
    x.append(i)
    y.append(inversion(0.001))
plt.plot(x,y)
plt.xlabel("Number of try")
plt.ylabel("Random number")
plt.title("Inversion plot at a=0.001")
plt.show()
#Acceptance Rejection technique
def accrej(a):
    b=0
    acc=0
    while b==0:
        u1 = random.uniform(0,1)
        u2 = random.uniform(0,1)
        fmax = 1/a
        fmin = 1
        t = fmin + (fmax-fmin)*u1
        if 1/(math.sin(t)**2 + a*math.cos(t)**2)>(u2*fmax):
            acc=t
            b = 1
    return acc
x = []
y = []
for i in range(10000):
    x.append(i)
    y.append(accrej(0.5))
plt.plot(x,y)
plt.xlabel("Number of try")
plt.ylabel("Random number")
plt.title("Acceptance-Rejection plot at a=0.5")
plt.show()
x = []
y = []
for i in range(10000):
    x.append(i)
    y.append(accrej(0.001))
plt.plot(x,y)
plt.xlabel("Number of try")
plt.ylabel("Random number")
plt.title("Acceptance-Rejection plot at a=0.001")
plt.show()
```
